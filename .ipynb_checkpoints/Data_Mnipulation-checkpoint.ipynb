12. mean


**Operations**

```python
df = pd.read_csv('datasets/heart.csv')
print(df.head()) #print the first 5 rows
print("\n-----------------print first 10 rows------------------\n")
print(df.head(10)) #print first 10 rows
print("\n----------------last 10 rows-------------------\n")
print(df.tail(10)) #last 10 rows
print("\n---------------Random 10 rows--------------------\n")
print(df.sample(10)) #Random 10 rows
print("\n---------------Get a summary of the coloumns data types and some other informations--------------------\n")
print(df.info()) #Get a summary of the coloumns data types and some other informations
print("\n---------------Returns the measure of dispersion of the dataset--------------------\n")
print(df.describe()) # Returns the measure of dispersion of the dataset
print("\n---------------Returns the number of rows and columns of the dataset--------------------\n")
print(df.shape) # Returns the number of rows and columns of the dataset
print("\n---------------Returns the relation between columns--------------------\n")
print(df.corr()) # Returns the relation between columns >0<=1 means both increases/decreases, <0 means one increase other decrease.
print("\n---------------Returns number of repetation of each value of the column--------------------\n")
print(df.value_counts('age')) # Returns number of repetation of each value of the column.
print("\n---------------Returns the max value for each column--------------------\n")
print(df.max()) # Returns the min value for each column.
print("\n---------------Returns the max value for each column--------------------\n")
print(df.min()) # Returns the min value for each column.
print("\n---------------Returns the avearage/mean value for each column--------------------\n")
print(df.mean()) # Returns the avearage/mean value for each column.
```
